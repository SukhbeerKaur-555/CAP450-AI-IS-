Pandas Series

```python
# Pandas series is like a column in a table
#It is a one-dimensional array holding data of any type.

import pandas as pd

a = [4, 5, 40]

myvar = pd.Series(a)

print(myvar)

# Accessing The value from series with default index lables
print(myvar[2])

#Accessing The value from series with given index lables

import pandas as pd

a = [4, 5, 40]

myvar = pd.Series(a, index = ["A", "B", "C"])

print(myvar)
print(myvar["C"])
```

** Pandas Dataframes**

```python
import pandas as pd

data = {

  "Sales": [20, 30, 90,100,150],
  "Profit": [550, 750, 2200,2500, 3500]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)
print(df)

#refer to the row index:
print(df.loc[0])
print(df.loc[[3,4]])

# Indexed dataframe
import pandas as pd

data = {

  "Sales": [20, 30, 90,100,150],
  "Profit": [550, 750, 2200,2500, 3500]
}

#load data into a DataFrame object:
df = pd.DataFrame(data, index=["Day1","Day2","Day3","Day4","Day5"])
print(df)


#refer to the named index:
print(df.loc["Day2"])


```
